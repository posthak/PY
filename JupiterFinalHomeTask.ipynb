```python
from sympy import *
from sympy.plotting import plot

x = symbols('x')
rangeMin, rangeMax = -5, 5
p1 = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30

def GetF(x):
    return round(-12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30)

def GetFMoreZero():
    a = []
    i = rangeMin
    while i <= rangeMax:
        s = round(GetF(i))
        if s > 0:
            a.append(round(i, 1))
        i += 0.5
    return a

def GetFLessZero():
    a = []
    i = rangeMin
    while i <= rangeMax:
        s = round(GetF(i))
        if s < 0:
            a.append(round(i, 1))
        i += 0.5
    return a

def GetRoots():
    a = []
    i=rangeMin
    while i <= rangeMax:
        s= GetF(i)
        if s==0:
            a.append(round(i,1))
        i+=0.0001
    return set(a)

def GetIntervalsUp():
    b = []
    i = rangeMin
    last = rangeMin
    while i <= rangeMax:
        s = round(GetF(i))
        if last<s:
            b.append(round(i, 1))
        last = s
        i += 0.5
    return b

def GetTopPoints():
    b = []
    i = rangeMin+0.5
    last = rangeMin
    while i <= rangeMax:
        s = round(GetF(i))
        if last > s < round(GetF(i+0.5)) or last < s > round(GetF(i+0.5)):
            b.append(round(i, 1))
        last = s
        i += 0.5
    return b

def GetIntervalsDown():
    b = []
    i = rangeMin
    last = rangeMin
    while i <= rangeMax:
        s = round(GetF(i))
        if last>s:
            b.append(round(i, 1))       
        last = s
        i += 0.5
    return b

def ShowGraph():
    x = symbols('x')
    p1 = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
    result = plot(p1, (x, rangeMin, rangeMax), show=false)
    result.show()

print('4. Построить график')
ShowGraph()

print('1. Определить корни')
print(GetRoots())

print('2. Найти интервалы, на которых функция возрастает')
print(GetIntervalsUp())

print('3. Найти интервалы, на которых функция убывает')
print(GetIntervalsDown())

print('5. Вычислить вершину')
print(GetTopPoints())

print('7. Промежутки, на котором f > 0')
print(GetFMoreZero())

print('6. Промежутки, на котором f < 0')
print(GetFLessZero())
```
